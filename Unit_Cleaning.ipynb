```python
# make all units lowercase
meats_no_nan['UNIT'] = meats_no_nan['UNIT'].str.lower()
```

```python
# creating the POUND dataframe (ending with '#')
pound = meats_no_nan[meats_no_nan['UNIT'].str.endswith('#')]
pound
# 9,328 rows 9422
```

```python
def pounds(x):
    """
    This function cleans the units which have '#'
    If there is a unit '10#' that is equivalent to 10 lbs
    If there is a unit '15/2#' that is equivalent to 30 lbs (15 bags of 2 lbs each) or '2/25#' is 2 bags of 25 lb each
    """
    if '/' not in x:
        return x[:-1]
    else:
        bags = float(re.findall('.*\/', x)[0][:-1])
        pounds_per = float(re.findall('\/.*#', x)[0][1:-1])
        return bags * pounds_per
```

```python
pound['Num_Pounds'] = pound['UNIT'].apply(pounds)
```

```python
# creating the DOZEN dataframe (ending with 'dz' or 'dzn')
dozen = meats_no_nan[meats_no_nan['UNIT'].str.endswith('dz') | meats_no_nan['UNIT'].str.endswith('dzn')]
dozen
# 2,037 rows (only contains CATEGORY == 'Egg') 2057
```

```python
def dozens(x):
    return float(re.findall('[0-9]{1,2}', x)[0])
```

```python
dozen['Num_Count'] = dozen['UNIT'].apply(dozens) * 12
```

```python
# creating the OUNCE dataframe (ending with 'z') AND take out those with 'd' for dozen
oz = meats_no_nan[meats_no_nan['UNIT'].str.endswith('z')] #5,341 rows
oz = oz[~oz['UNIT'].str.contains('d')] #3,306 rows # takes out 'dz'
```

```python
ounce = oz[~oz['UNIT'].str.contains('o')] 
ounce
#3,301 rows (only took out 5 rows) 3345
# '0/4 oz' and '/30 oz' to be dealt with afterwards
```

```python
def ounces(x):
    if '/' not in x:
        return x[:-1]
    else:
        bags = float(re.findall('.*\/', x)[0][:-1])
        pounds_per = float(re.findall('\/.*z', x)[0][1:-1])
        return bags * pounds_per
```

```python
# convert to pounds
# 1 oz = 0.0625 lb
ounce['Num_Pounds'] = (ounce['UNIT'].apply(ounces)) * 0.0625
```

```python
# Check for '0/4 oz' and '/30 oz'
oz_04 = oz[oz['UNIT'] == '0/4 oz']
oz_30 = oz[oz['UNIT'] == '/30 oz']
oz_only = pd.concat([oz_04, oz_30], axis=0, ignore_index=True)
# 5 rows
```

```python
def oz_cleaner(x):
    # this was an error, pack size is 5/30oz, each case 5 bags of 30 ounces each bag
    if x == '/30 oz':
        return 150
    # this is 80/4oz, 80 patties 4 oz each per case
    elif x == '0/4 oz':
        return 320
    else:
        return None
```

```python
# convert to pounds
# 1 oz = 0.0625 lb
oz_only['Num_Pounds'] = (oz_only['UNIT'].apply(oz_cleaner)) * 0.0625
```

```python
lb_only = meats_no_nan[meats_no_nan['UNIT'].str.endswith('lb')]
lb_only
# 5,493 rows 5508
```

```python
def lbs(x):
    """
    This function cleans the units which have 'lb'
    If there is a unit '10lb' or '10/lb' that is equivalent to 10 lbs ('lb' is equivalent to 1 lb)
    If there is a unit '15/2lb' that is equivalent to 30 lbs (15 bags of 2 lbs each)
    """
    if '/' not in x:
        x = x[:-2]
        if len(x) == 0:
            return 1
        else:
            return float(re.findall('[^\s]+', x)[0])
    else:
        bags = float(re.findall('.*\/', x)[0][:-1])
        if '/lb' in x:
            return bags
        else:
            pounds_per = float(re.findall('\/.*lb', x)[0][1:-2])
            return bags * pounds_per
```

```python
lb_only['Num_Pounds'] = lb_only['UNIT'].apply(lbs)
```

```python
# dataframe COUNT which has UNIT ending with 'ct'
count = meats_no_nan[meats_no_nan['UNIT'].str.endswith('ct')]
count
# 1,608 rows 1611
```

```python
def counts(x):
    """
    This function cleans the units which have 'ct'
    If there is a unit '10ct' or '10/ct' that is equivalent to 10 ct
    If there is a unit '15/2 ct' that is equivalent to 30 ct (15 bags of 2 ct each)
    """
    if '/' not in x:
        x = x[:-2]
        if len(x) == 0:
            return 1
        else:
            return float(re.findall('[^\s]+', x)[0])
    else:
        bags = float(re.findall('.*\/', x)[0][:-1])
        if '/ct' in x:
            return bags
        else:
            pounds_per = float(re.findall('\/.*ct', x)[0][1:-2])
            return bags * pounds_per
```

```python
count['Num_Count'] = count['UNIT'].apply(counts)
```

```python
# dataframe 'CASE' those ending with 'cs', 'case', or 'cas'
case = meats_no_nan[meats_no_nan['UNIT'].str.endswith('cs') | meats_no_nan['UNIT'].str.endswith('case') | meats_no_nan['UNIT'].str.endswith('cas')]
case
# 3,294 rows 3309
```

```python
def cases(x):
    """
    This function cleans the units which have 'ct'
    If there is a unit '10ct' or '10/ct' that is equivalent to 10 ct
    If there is a unit '15/2 ct' that is equivalent to 30 ct (15 bags of 2 ct each)
    """
    if x == 'case':
        return 1
    elif '/' not in x and '#' not in x:
        x = x[:-2]
        if len(x) == 0:
            return 1
        else:
            return float(re.findall('[^\s]+', x)[0])
    else:
        bags = float(re.findall('.*[/#]', x)[0][:-1])
        if ('/cs' in x) or ('#cs' in x) or ('/cas' in x):
            return bags
        else:
            pounds_per = float(re.findall('[/#].*[cs|cas]', x)[0][1:-2])
            return bags * pounds_per
```

```python
case['Num_Cases'] = case['UNIT'].apply(cases)
```

```python
weird_ct = meats_no_nan[(meats_no_nan['UNIT'] == '2/12 c') | (meats_no_nan['UNIT'] == '16/2c')]
weird_ct
# 100 rows 100
```

```python
def weird_ct_def(x):
    if x == '2/12 c':
        return 24
    elif x == '16/2c':
        return 32
    else:
        return None
```

```python
weird_ct['Num_Count'] = weird_ct['UNIT'].apply(weird_ct_def)
```

```python
weird_lb = meats_no_nan[(meats_no_nan['UNIT'] == '07/1.5') | (meats_no_nan['UNIT'] == '4/2.5') | 
             (meats_no_nan['UNIT'] == '6/2.5') | (meats_no_nan['UNIT'] == '2/16.3') | 
             (meats_no_nan['UNIT'] == '0/5.33') | (meats_no_nan['UNIT'] == '8/7.3') | 
             (meats_no_nan['UNIT'] == '/27.09') | (meats_no_nan['UNIT'] == 'each') |
             (meats_no_nan['UNIT'] == '2/15 o') | (meats_no_nan['UNIT'] == '2/12 o') | 
             (meats_no_nan['UNIT'] == '6/11 o') | (meats_no_nan['UNIT'] == '0/30.5') | 
             (meats_no_nan['UNIT'] == '0/16 o') | (meats_no_nan['UNIT'] == '/12/3.5') | 
             (meats_no_nan['UNIT'] == '/#10 c') | (meats_no_nan['UNIT'] == '/10.10') | 
             (meats_no_nan['UNIT'] == '6/#10')]
weird_lb
# 2,150 rows 2155
```

```python
def weird_lb_def(x):
    # 107/1.5oz â€“ 107 patties in a case, 1.5oz per patty
    if x == '07/1.5':
        return (107 * 1.5) * 0.0625
    # 4 packages per case, 2.5lb per package
    elif x == '4/2.5':
        return 4 * 2.5
    # 6 bags per case, 2.5lb per bag
    elif x == '6/2.5':
        return 6 * 2.5
    # 12/16.3oz â€“ 12 jars per case, 16.3oz per jar
    elif x == '2/16.3':
        return (12 * 16.3) * 0.0625
    # 30/5.33oz â€“ 30 per case, 5.33oz each
    elif x == '0/5.33':
        return (30 * 5.33) * 0.0625
    # 48/7.3oz â€“ 48 per case, 7.3oz each
    elif x == '8/7.3':
        return (48 * 7.3) * 0.0625
    # 6/30oz â€“ 6 bags per case, 30oz per bag
    elif x == '/27.09':
        return (6 * 30) * 0.0625
    # 6/#10can â€“ 6 cans per case, 109oz per can
    elif x == 'each':
        return (6 * 109) * 0.0625
    elif x == '2/15 o':
        return (2 * 15) * 0.0625
    elif x == '2/12 o':
        return (2 * 12) * 0.0625
    elif x == '6/11 o':
        return (6 * 11) * 0.0625
    # this was an error, pack size is 5/30oz, each case 5 bags of 30 ounces each bag
    elif x == '0/30.5':
        return 150 * 0.0625
    # this is 20/16oz, so 20, 16 ounce bags per case
    elif x == '0/16 o':
        return (20 * 16) * 0.0625
    # this is 4/12/3.5oz, each case has 4 bags, each bag has 12 patties, each patty is 3.5oz
    elif x == '/12/3.5':
        return (4 * 12 * 3.5) * 0.0625
    # 6/#10can â€“ 6 cans per case, 109oz per can
    elif x == '/#10 c':
        return (6 * 109) * 0.0625
    # this was cut off, it is 8/10.10oz, so each case is 8 bags of 10.10 ounces each bag
    elif x == '/10.10':
        return (8 * 10.10) * 0.0625
    # 6/#10can â€“ 6 cans per case, 109oz per can
    elif x == '6/#10':
        return (6 * 109) * 0.0625
    else:
        return None
```

```python
weird_lb['Num_Pounds'] = weird_lb['UNIT'].apply(weird_lb_def)
```
